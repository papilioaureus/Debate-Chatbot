```python
def save_configurations(configs, filename):
    """
    Save configurations to a file using pickle.

    Args:
    configs (dict): Configuration dictionary to save.
    filename (str): The path to the file where the configurations will be saved.
    """
    with open(filename, 'wb') as file:
        pickle.dump(configs, file)
    print(f"Configurations saved successfully to {filename}")
```

```python
def load_configurations(filename):
    """
    Load configurations from a file using pickle.

    Args:
    filename (str): The path to the file from which the configurations will be loaded.

    Returns:
    dict: The loaded configurations.
    """
    with open(filename, 'rb') as file:
        return pickle.load(file)
```

```python
def recreate_model_from_configs(configs):
    documents = SimpleDirectoryReader(configs['documents_directory']).load_data()

    llm = LlamaCPP(
        model_url=configs['llama_model_url'],
        temperature=configs['llama_temperature'],
        max_new_tokens=configs['llama_max_new_tokens'],
        context_window=configs['llama_context_window'],
        generate_kwargs=configs['llama_generate_kwargs'],
        model_kwargs=configs['llama_model_kwargs'],
        messages_to_prompt=messages_to_prompt,
        completion_to_prompt=completion_to_prompt,
        verbose=True,
    )

    embed_model = LangchainEmbedding(
      HuggingFaceEmbeddings(model_name=configs['embedding_model_name'])
    )

    service_context = ServiceContext.from_defaults(
        chunk_size=configs['service_context_chunk_size'],
        llm=llm,
        embed_model=embed_model
    )

    index = VectorStoreIndex.from_documents(documents, service_context=service_context)
    return index

# Save the configurations with a specified filename
configs_filename = '/Users/cristina/Desktop/my_model_configs.pkl'
save_configurations(configs, configs_filename)

```
